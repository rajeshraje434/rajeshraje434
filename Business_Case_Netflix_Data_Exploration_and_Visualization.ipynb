# 1. How has the number of movies released per year changed over the last 20-30 years?

```python
# Convert 'release_year' to datetime
netflix_data['release_year'] = pd.to_datetime(netflix_data['release_year'], format='%Y', errors='coerce')
```

```python
# Filter data for the last 30 years
current_year = pd.Timestamp.now().year
start_year = current_year - 30
filtered_data = netflix_data[(netflix_data['release_year'].dt.year >= start_year) & (netflix_data['release_year'].dt.year <= current_year)]
```

```python
# Filter for movies
movies_data = filtered_data[filtered_data['type'] == 'Movie']
```

```python
# Group data by release year and count the number of movies per year
movies_per_year = movies_data.groupby(movies_data['release_year'].dt.year)['show_id'].count()
```

```python
# Create a line plot to visualize the trend
plt.figure(figsize=(12, 6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o', linestyle='-')
plt.title('Number of Movies Released per Year (Last 30 Years)')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

## 3. What is the best time to launch a TV show?

```python
# Filter for TV shows
tv_shows_data = netflix_data[netflix_data['type'] == 'TV Show']
```

```python
# Convert 'date_added' column to datetime
tv_shows_data['date_added'] = pd.to_datetime(tv_shows_data['date_added'])
```

```python
# Extract the year and month of addition
tv_shows_data['year_added'] = tv_shows_data['date_added'].dt.year
tv_shows_data['month_added'] = tv_shows_data['date_added'].dt.month
```

```python
# Group data by year and month to count the number of TV shows added
tv_shows_added_per_month = tv_shows_data.groupby(['year_added', 'month_added']).size()
```

```python
# Create a line plot to visualize the trend
plt.figure(figsize=(12, 6))
tv_shows_added_per_month.plot(marker='o', linestyle='-', color='blue')
plt.title('Number of TV Shows Added to Netflix Over Time')
plt.xlabel('Year and Month')
plt.ylabel('Number of TV Shows Added')
plt.grid(True)
plt.show()
```

# 4. Analysis of actors/directors of different types of shows/movies

```python
# Separate data into Movies and TV Shows
movies_data = netflix_data[netflix_data['type'] == 'Movie']
tv_shows_data = netflix_data[netflix_data['type'] == 'TV Show']
```

```python
# Analyze actors for Movies
movie_cast = movies_data['cast'].str.split(', ').explode()
top_movie_actors = movie_cast.value_counts().head(10)
```

```python
# Analyze directors for Movies
top_movie_directors = movies_data['director'].value_counts().head(10)
```

```python
# Analyze actors for TV Shows
tv_show_cast = tv_shows_data['cast'].str.split(', ').explode()
top_tv_show_actors = tv_show_cast.value_counts().head(10)
```

```python
# Analyze directors for TV Shows
top_tv_show_directors = tv_shows_data['director'].value_counts().head(10)
```

```python
# Plot the results
plt.figure(figsize=(16, 10))
plt.subplot(2, 2, 1)
sns.barplot(x=top_movie_actors.values, y=top_movie_actors.index, palette='viridis')
plt.title('Top Movie Actors')

plt.subplot(2, 2, 2)
sns.barplot(x=top_movie_directors.values, y=top_movie_directors.index, palette='viridis')
plt.title('Top Movie Directors')

plt.subplot(2, 2, 3)
sns.barplot(x=top_tv_show_actors.values, y=top_tv_show_actors.index, palette='viridis')
plt.title('Top TV Show Actors')

plt.subplot(2, 2, 4)
sns.barplot(x=top_tv_show_directors.values, y=top_tv_show_directors.index, palette='viridis')
plt.title('Top TV Show Directors')

plt.tight_layout()
plt.show()
```

## 5. Does Netflix has more focus on TV Shows than movies in recent years?

```python
# Convert 'date_added' column to datetime
netflix_data['date_added'] = pd.to_datetime(netflix_data['date_added'])
```

```python
# Extract the year from the 'date_added' column
netflix_data['year_added'] = netflix_data['date_added'].dt.year
```

```python
# Filter data for the last 5 years (you can adjust the timeframe)
current_year = pd.Timestamp.now().year
start_year = current_year - 5
filtered_data = netflix_data[(netflix_data['year_added'] >= start_year) & (netflix_data['year_added'] <= current_year)]
```

```python
# Count the number of TV shows and movies added each year
tv_shows_added_per_year = filtered_data[filtered_data['type'] == 'TV Show'].groupby('year_added')['show_id'].count()
movies_added_per_year = filtered_data[filtered_data['type'] == 'Movie'].groupby('year_added')['show_id'].count()
```

```python
# Create a line plot to visualize the trend
plt.figure(figsize=(12, 6))
plt.plot(tv_shows_added_per_year.index, tv_shows_added_per_year.values, label='TV Shows', marker='o', linestyle='-')
plt.plot(movies_added_per_year.index, movies_added_per_year.values, label='Movies', marker='o', linestyle='-')
plt.title('Number of TV Shows vs. Movies Added to Netflix (Last 5 Years)')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.legend()
plt.grid(True)
plt.show()
```

## 6. Understanding what content is available in different countries

```python
# Group data by country and count the number of titles available
titles_by_country = netflix_data.groupby('country')['show_id'].count().reset_index()
```

```python
# Sort the data by the number of titles in descending order
titles_by_country_sorted = titles_by_country.sort_values(by='show_id', ascending=False)
```

```python
# Filter for the top N countries (you can adjust N)
top_countries = 10
top_countries_data = titles_by_country_sorted.head(top_countries)
```

```python
# Create a bar plot to visualize the top countries by the number of titles
plt.figure(figsize=(12, 6))
plt.barh(top_countries_data['country'], top_countries_data['show_id'], color='skyblue')
plt.title('Top Countries with the Most Titles on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.gca().invert_yaxis()  # Reverse the y-axis to show the highest count at the top
plt.show()
```
